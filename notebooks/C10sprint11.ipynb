```python
# Simulated Data
price_comparison = pd.DataFrame({
    'category': ['Electronics', 'Clothing', 'Home', 'Toys'],
    'price_diff': [120, -30, 50, -10]  # Positive = We are costlier
})

plt.figure(figsize=(10, 6))
colors = price_comparison['price_diff'].apply(lambda x: 'green' if x < 0 else 'red')
sns.barplot(data=price_comparison, x='price_diff', y='category', palette=colors)
plt.axvline(0, color='black', linewidth=0.8)
plt.title("Price Difference with Competitors by Category", fontsize=14)
plt.xlabel("Price Difference (Our Price - Competitor Price)")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
promo_effect = pd.DataFrame({
    'category': ['Electronics', 'Clothing', 'Home', 'Toys'],
    'our_discount': [10, 15, 12, 8],
    'competitor_discount': [15, 20, 18, 10]
})

x = np.arange(len(promo_effect['category']))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, promo_effect['our_discount'], width, label='Our Discount', color='skyblue')
plt.bar(x + width/2, promo_effect['competitor_discount'], width, label='Competitor Discount', color='orange')

plt.xlabel("Category")
plt.ylabel("Discount (%)")
plt.title("Discount Strategies: Ours vs Competitors")
plt.xticks(ticks=x, labels=promo_effect['category'])
plt.legend()
plt.tight_layout()
plt.show()
```
