```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\AKASH KRITHIK\Desktop\Amazon Sales Data Analysis\Amazon data.csv"
df = pd.read_csv(file_path)

# Display basic info before cleaning
print("Initial Data Info:")
df.info()
print("\nInitial Missing Values:\n", df.isnull().sum())
print(df.head())
print(df.info())
df["rating_count"] = df["rating_count"].astype(str).str.replace(",", "").replace("nan", "0").astype(float).astype(int)

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
# Remove % symbol and convert to float
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)

# Convert 'price' column to numeric if it exists
# Remove commas and convert to float
df['actual_price'] = df['actual_price'].astype(str).str.replace(',', '').astype(float)
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)

if 'price' in df.columns:
    df['price'] = pd.to_numeric(df['price'], errors='coerce')
if 'revenue' in df.columns:
    df.drop(columns=['revenue'], inplace=True)



# Calculate Revenue
df['revenue'] = df['actual_price'] * df['rating_count']
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')  # Fix string values

df['user_id'] = df['user_id'].str.split(',')
df = df.explode('user_id').reset_index(drop=True)


# Drop duplicate rows
df = df.drop_duplicates().copy()

# Clean numeric columns to remove commas and convert to float
df['actual_price'] = pd.to_numeric(df['actual_price'].astype(str).str.replace(',', ''), errors='coerce')
df['rating'] = pd.to_numeric(df['rating'].astype(str).str.replace(',', ''), errors='coerce')

# Drop rows with NaN values in the cleaned columns
df = df.dropna(subset=['actual_price', 'rating'])

# Handle missing values
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode().iloc[0])  # Fill categorical columns with mode
    else:
        df[col] = df[col].fillna(df[col].median())  # Fill numerical columns with median

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Extract top-level and leaf-level categories from the 'category' column
if 'category' in df.columns:
    df['top_category'] = df['category'].apply(lambda x: str(x).split('|')[0])
    df['leaf_category'] = df['category'].apply(lambda x: str(x).split('|')[-1])


# Convert 'price' column to numeric if it exists
if 'price' in df.columns:
    df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Display final info after cleaning
print("\nFinal Data Info:")
df.info()
print("\nFinal Missing Values:\n", df.isnull().sum())

# Save the cleaned dataset
cleaned_file_path = r"C:\Users\AKASH KRITHIK\Desktop\Amazon Sales Data Analysis\cleaned_amazon_data.csv"
df.to_csv(cleaned_file_path, index=False)
print("\nCleaned data saved successfully at:", cleaned_file_path)
```
