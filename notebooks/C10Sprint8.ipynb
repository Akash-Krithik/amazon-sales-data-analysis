```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataframe (replace with your actual data)
file_path = r"C:\Users\AKASH KRITHIK\Desktop\Amazon Sales Data Analysis\cleaned_amazon_data.csv"
df = pd.read_csv(file_path)
# print(df.columns)


# Question 1: Top-performing products based on revenue and ratings
top_products = df.sort_values(by=['revenue', 'rating'], ascending=False).head(10)
print("Top 10 Products Based on Revenue and Ratings:")
print(top_products[['product_name', 'revenue', 'rating']])


# Question 2: Rating Count vs Revenue and Actual Price vs Revenue
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='rating_count', y='revenue', hue='rating', palette='viridis')
plt.title("Rating Count vs Revenue")
plt.xlabel("Rating Count")
plt.ylabel("Revenue")

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='actual_price', y='revenue', hue='rating', palette='coolwarm')
plt.title("Actual Price vs Revenue")
plt.xlabel("Actual Price")
plt.ylabel("Revenue")

plt.tight_layout()
plt.show()

```
