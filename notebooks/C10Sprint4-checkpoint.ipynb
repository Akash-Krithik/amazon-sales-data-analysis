```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix, classification_report
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample

# Load Data
file_path = r"C:\Users\AKASH KRITHIK\Desktop\Amazon Sales Data Analysis\cleaned_amazon_data.csv"
df = pd.read_csv(file_path)

# Convert 'discount_percentage' from string to float
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=True).astype(float)

# Clean numeric columns
numeric_cols = ['discounted_price', 'actual_price', 'rating', 'rating_count', 'discount_percentage', 'revenue']
df[numeric_cols] = df[numeric_cols].replace(',', '', regex=True).astype(float)

# Remove NaNs and Infs
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# -------------------------------
# CREATE TARGET for CLASSIFICATION before log transform
# -------------------------------
threshold = df['revenue'].quantile(0.70)
df['high_revenue'] = (df['revenue'] > threshold).astype(int)
print("‚úÖ Class distribution:\n", df['high_revenue'].value_counts())

# -------------------------------
# LOG TRANSFORM for REGRESSION only
# -------------------------------
upper_limit = df['revenue'].quantile(0.99)
df['revenue'] = np.where(df['revenue'] > upper_limit, upper_limit, df['revenue'])
df['revenue'] = np.log1p(df['revenue'])

# -------------------------------
# LINEAR REGRESSION
# -------------------------------
features = ['discounted_price', 'actual_price', 'rating', 'rating_count', 'discount_percentage']
X = df[features]
y = df['revenue']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train_lr, y_train_lr)
y_pred_lr = model.predict(X_test_lr)

print("\nüìä Linear Regression Evaluation:")
print("Mean Squared Error:", mean_squared_error(y_test_lr, y_pred_lr))
print("R¬≤ Score:", r2_score(y_test_lr, y_pred_lr))

plt.figure(figsize=(8, 6))
plt.scatter(y_test_lr, y_pred_lr, alpha=0.6)
plt.title('Actual vs Predicted Revenue')
plt.xlabel('Actual Revenue')
plt.ylabel('Predicted Revenue')
plt.show()

# -------------------------------
# LOGISTIC REGRESSION (Balanced)
# -------------------------------
# Balance the data
majority = df[df['high_revenue'] == 0]
minority = df[df['high_revenue'] == 1]

majority_downsampled = resample(
    majority,
    replace=False,
    n_samples=len(minority),
    random_state=42
)

balanced_df = pd.concat([majority_downsampled, minority]).sample(frac=1, random_state=42)

X_bal = balanced_df[features]
y_bal = balanced_df['high_revenue']
X_bal_scaled = scaler.fit_transform(X_bal)

X_train, X_test, y_train, y_test = train_test_split(X_bal_scaled, y_bal, test_size=0.2, random_state=42, stratify=y_bal)

print("\n‚úÖ Training class distribution:\n", y_train.value_counts())

clf = LogisticRegression(max_iter=2000, solver='lbfgs', class_weight='balanced')
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("\nüîç Logistic Regression Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# -------------------------------
# K-MEANS CLUSTERING
# -------------------------------
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discounted_price', y='revenue', hue='cluster', palette='viridis')
plt.title('Customer Segmentation based on Revenue and Discounted Price')
plt.xlabel('Discounted Price')
plt.ylabel('Log(Revenue)')
plt.show()
```
